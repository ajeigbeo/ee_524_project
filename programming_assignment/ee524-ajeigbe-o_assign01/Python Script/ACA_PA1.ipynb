```python
#load required pacakages
import pandas as pd
```

Converting a Text-File into a Dataframe

```python
#create empty dataframe
fl1 = pd.DataFrame()

# imports data
file1 = "./Data/runfl1.txt"
file2 = "./Data//runfl2.txt"
file3 = "./Data//runfl3.txt"

files = file1, file2, file3

for tfile in files:

    with open(tfile) as f:

        for i in range(24):

            line = f.readline()
            line = line.strip()
            temp = line.split()
            
            name = temp[1]
            coretype = temp[5]
            corethreads = temp[7]
            corenum = temp[11]

            f.readline()
            f.readline()
            f.readline()
            f.readline()
            f.readline()
            instuctions_val = f.readline().split()[0]
            cycles_val = f.readline().split()[0]
            mis_pred = f.readline().split()[0]
            l1d_c = f.readline().split()[0]
            l1i_c = f.readline().split()[0]
            l2d_c = f.readline().split()[0]
            f.readline()
            exec_time = f.readline().split()[0]
            f.readline()
            f.readline()
            f.readline()
            f.readline()
            f.readline()

            tempdf = pd.DataFrame()
            tempdf['name'] = [name]
            tempdf['coretype'] = coretype
            tempdf['corethreads'] = corethreads
            tempdf['corenum'] = corenum
            tempdf['instuctions_val'] = instuctions_val
            tempdf['cycles_val'] = cycles_val
            tempdf['mis_pred'] = mis_pred
            tempdf['l1d_c'] = l1d_c
            tempdf['l1i_c'] = l1i_c
            tempdf['l2d_c'] = l2d_c
            tempdf['exec_time'] = exec_time
            tempdf['filename'] = tfile

            fl1 = pd.concat([fl1, tempdf], ignore_index = True)


    
```

```python
fl1.to_csv('fl.csv')
```
